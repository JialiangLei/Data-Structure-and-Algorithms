### 2.

```python
import timeit
import random

for i in range(1000,100001,2000):
    
    t = timeit.Timer("x.get(random.randrange(%d))" % i,
                     "from __main__ import random, x")
    x = {j:None for j in range(i)}
    get_time = t.timeit(number = 1000)

    t = timeit.Timer("x[random.randrange(%d)] = random.randrange(%d)" %(i,i),
                     "from __main__ import random, x")
    int_time = t.timeit(number = 1000)

    print("%d, %10.3f, %10.3f" % (i, get_time, int_time))
```

### 3.

```python
import timeit
import random

for i in range(1000,100001,2000):

    t = timeit.Timer("del x[random.randrange(%d)]" % i,
                     "from __main__ import random, x")
    x = {j:None for j in range(i)}
    dict_time = t.timeit(number = 1)

    t = timeit.Timer("del x[random.randrange(%d)]" % i,
                     "from __main__ import random, x")
    x = list(range(i))
    lst_time = t.timeit(number = 1)

    print("%d, %10.3f, %10.3f" % (i, dict_time, lst_time))
```
