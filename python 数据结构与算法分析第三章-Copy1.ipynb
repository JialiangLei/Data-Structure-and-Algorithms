```python
class Node:
    def __init__(self, data):
        self.prev = None
        self.data = data
        self.next = None

class DoubleLinkList:
    def __init__(self):
        self.head = None
        self.size = 0

    def is_empty(self):
        return self.size == 0

    def get_size(self):
        return self.size

    def items(self):
        lst = []
        node = self.head
        while node != None:
            lst.append(node.data)
            node = node.next
        return lst

    def add(self, data):
        node = Node(data)
        if self.is_empty():
            self.head = node
        else:
            self.head.prev = node
            node.next = self.head
            self.head = node
        self.size += 1

    def append(self, data):
        node = Node(data)
        if self.head != None:
            cur = self.head
            while cur.next != None:
                cur = cur.next
            node.prev = cur
            cur.next = node
        else:
            self.head = node
        self.size += 1

    def search(self, data):
        node = self.head
        while True:
            if node.data == data:
                return True
            node = node.next
        return False

l = DoubleLinkList()
l.add(1)
print(l.items())
l.add(2)
print(l.items())
l.append(3)
print(l.items())
print(l.search(2))
```
